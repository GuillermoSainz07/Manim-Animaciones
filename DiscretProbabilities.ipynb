### Distribuciones discretas

```python
%%manim -ql -v WARNING Discret_Probability_2

class Discret_Probability_2(ThreeDScene):
    def construct(self):
        axes = ThreeDAxes(x_range=[0,6],
                         y_range=[0,6],
                         z_range=[0,5]).scale(0.6).to_edge(LEFT, buff=1)

        lines_x = VGroup()
        lines_y = VGroup()

        for i in range(1,6):
            lines_x.add(Line(axes.c2p(i,0,0),
                           axes.c2p(i,5,0),
                            stroke_width=1))

            lines_y.add(Line(axes.c2p(0,i,0),
                            axes.c2p(5,i,0),
                            stroke_width=1))

        dot_xy = always_redraw(lambda: Dot3D(axes.c2p(1,1,0), color=PURPLE, radius=0.1))

        label_x = axes.get_x_axis_label(MathTex('X'))
        label_y = axes.get_y_axis_label(MathTex('Y'))
        label_z = axes.get_z_axis_label(MathTex('f(X,Y)'))
        labels = VGroup(label_x, label_y, label_z).set_color(BLUE)

        title = Text('Probabilidades Conjuntas Discretas').scale(0.4).set_color(YELLOW).to_edge(UR).shift(LEFT)
        explain = MathTex('&\\text{Cuando el espacio objetivo es discreto, podemos imaginar la distribucion de}', '\\\\',
                          '&\\text{probabilidad de multiples variables aleatorias como el llenado de una matriz}', '\\\\',
                          '&\\text{multidimensional. En este caso la funcion de probabilidad se define como}', '\\\\',
                          '&\\ P(X_1 = x_1, ... , X_n = x_n)','\\\\').scale(0.4).next_to(title, DOWN).shift(LEFT*0.2)


        explain_2 = MathTex('&\\text{Para un vector aleatorio}','X_1, ... , X_n','\\text{la probabilidad de}','\\\\',
                          '&\\ X_1 = x_1','\\text{y}','X_2=x_2','\\text{y}','...','\\text{y}','X_n = x_n','\\\\',
                          '&\\text{se escribe como}','p(x_1,...,x_n)','\\text{y se le llamada probabilidad conjunta}','\\\\',
                          '&\\ P_{X_1,...,X_n}(x_1,...,x_n) = P(X_1=x_1,...,X_n=x_n)').scale(0.4).next_to(explain, DOWN).shift(LEFT*0.5)




        grafica = VGroup(axes, lines_x, lines_y, *labels)


        self.play(Create(grafica))
        self.play(Write(title))
        self.play(FadeIn(explain))
        self.add_fixed_in_frame_mobjects(title)
        self.add_fixed_in_frame_mobjects(explain)
        self.play(Create(dot_xy))

        self.move_camera(zoom=0.8)
        self.move_camera(phi=70*DEGREES)
        self.move_camera(theta=-60*DEGREES)

        grafica = VGroup(axes, lines_x, lines_y, *labels)
        self.play(grafica.animate.move_to(DOWN*0.01 + LEFT*4))

        self.add_fixed_in_frame_mobjects(explain_2)
        self.wait()

        text_fx_1 = VGroup()

        for i in range(1,6):

          # Modificacion variable aleatoria Y
          self.play(Create(Dot3D(axes.c2p(1,i,0), color=PURPLE, radius=0.1)))

          self.play(Create(DashedLine(axes.c2p(1,i,0),
                              axes.c2p(1,i,3))))

          self.play(Create(Dot3D(axes.c2p(1,i,3), color=RED, radius=0.1)))
          text = MathTex(f'f(X=x_1, Y = y_{i})').scale(0.4).rotate(PI/2, axis=RIGHT).move_to(axes.c2p(1,i,3.4))
          text_fx_1.add(text)
          self.play(Write(text))

        self.play(FadeOut(text_fx_1))

        text_fx_2 = VGroup()
        for i in range(1,6):
          # Modificacion variable aleatoria X
          self.play(Create(Dot3D(axes.c2p(i,1,0), color=YELLOW, radius=0.1)))

          self.play(Create(DashedLine(axes.c2p(i,1,0),
                              axes.c2p(i,1,3))))

          self.play(Create(Dot3D(axes.c2p(i,1,3), color=BLUE, radius=0.1)))
          text_3 = MathTex(f'f(X=x_{i}, Y = y_1)').scale(0.4).rotate(PI/2, axis=RIGHT).move_to(axes.c2p(i,1,3.4))
          text_fx_2.add(text_3)

          self.play(Write(text_3))

        self.play(FadeOut(text_fx_2))
        self.wait(3)
```
