```python
import numpy as np
import scipy.stats as st
```

```python
%%manim -qh -v WARNING Gaussian_Density_2
class Gaussian_Density_2(Scene):
    def construct(self):
        axes = Axes(x_range=[-3,3],
                   y_range=[0,0.7])

        x = np.linspace(-3,3, 10)

        funcion = lambda p: st.norm(0,1).pdf(x)

        plot = axes.plot(lambda x: st.norm(0,1).pdf(x),).set_color(PURPLE).set_stroke(width=5)


        dot_x = VGroup()
        dashes = VGroup()
        x_labels = VGroup()
        dot_y = VGroup()

        c = 0
        n = 0
        y_label = axes.get_y_axis_label(MathTex('f(x)')).scale(0.6)
        self.add(axes,y_label)
        for i in x:
            dot_x = Dot().move_to(axes.c2p(i,0)).set_color(BLUE)
            dot_y = Dot().move_to(axes.c2p(i, plot.underlying_function(i))).set_color(YELLOW)
            self.play(Write(MathTex(f'x_{c}').scale(0.4).next_to(dot_x, DOWN)))
            self.play(Create(dot_x))
            self.play(Create(DashedLine(axes.c2p(i,0),
                                 axes.c2p(i, plot.underlying_function(i)))))

            self.play(Create(axes.get_graph_label(x_val=i,
                                              graph=plot,
                                              label=MathTex(f'f(x_{n})').scale(0.4),
                                              direction=UP).set_color(WHITE)))
            self.play(Create(dot_y))

            n+=1
            c+=1

        area = axes.get_riemann_rectangles(x_range=[-3,3],
                             dx=0.05,
                             graph=plot,
                             color=BLUE,
                             fill_opacity=0.4)

        integral = ValueTracker(0)
        text_integral = MathTex('\int^{\infty}_{-\infty} f(x) dx = ').scale(0.6).to_edge(UR).shift(DOWN + LEFT)

        integral_n = always_redraw(lambda: DecimalNumber(integral.get_value()).scale(0.4).next_to(text_integral).set_color(YELLOW))

        text_density = MathTex('f(x) = \\frac1{\sqrt{2\pi\sigma^2}}{exp}^\\frac{x-\mu}{\sigma}').scale(0.6).to_edge(UL).shift(DOWN + RIGHT)

        self.play(Write(text_integral),Write(text_density) ,Write(integral_n))
        self.play(Create(area), integral.animate.set_value(1), run_time=3)
        self.play(Create(plot))
        self.wait()
```
